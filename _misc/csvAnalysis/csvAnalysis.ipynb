## csv Analysis

This simply plots a frequency histogram. 

```python
data_file = 'data/text2textGen.csv'
selected_column = 'task'
```

```python
# IMPORTS
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme()
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
try:
    df = pd.read_csv(data_file, index_col=0)
except Exception as e:
    print(f"Error reading {data_file}: {e}")

fig, ax = plt.subplots(figsize=(10, 6))
counts = df[selected_column].value_counts()
ax.bar(counts.index.str.capitalize(), counts.values, color='blue', edgecolor='black', alpha=0.7)
ax.set_title(f'Count of {selected_column} in \'{data_file[5:-4]}\'', fontsize=16)
ax.set_xlabel(selected_column.capitalize(), fontsize=14)
ax.set_ylabel('Count', fontsize=14)
ax.tick_params(axis='x', rotation=300)

counts
```
